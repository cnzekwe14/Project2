```python
path2 = (r'C:\Users\Christina\Desktop\Project2\olympics_medals_country_wise.csv')
olympic_df = pd.read_csv(path2)
olympic_df
```

```python
char = ['(',')']

#removed the parantheses from the ioc_code
for x in char:
    olympic_df['ioc_code ']=olympic_df['ioc_code '].str.replace(x,'')
olympic_df

list(olympic_df.columns)
```

```python
#replace countries in olympic_df with Country to merge with HDI_df
#replace total_total with total_medals
#get rid of the extra space from ioc_code
olympic_df.rename(columns={'countries ':'Country','total_total ': 'total_medals','ioc_code ':'ioc_code'},inplace = True)
olympic_df

```

```python
#convert columns to integer
char2 = [',']
for x in char2:
    olympic_df['summer_gold']=olympic_df['summer_gold'].str.replace(x,'')
olympic_df

olympic_df['summer_gold'] = pd.to_numeric(olympic_df['summer_gold'])
olympic_df

```
